```python
txt = input("請輸入第一個字串")
strList = []
strList.append(txt)
msg = txt
inputMsg = f"請輸入-{txt[-1]}-開始的字串"
count = 0
while count < 5:
    tmpTxt = input(inputMsg)
    if tmpTxt[0] == txt[-1]:
        txt = tmpTxt
        strList.append(tmpTxt)
        msg = "-".join(strList)
        inputMsg = f"請輸入-{txt[-1]}-開始的字串"
    else:
        inputMsg = f"請再次輸入-{txt[-1]}-開始的字串"
        count += 1
    print("上一個字串是:",msg)    
```

## Dictionary

資料結構是由 key:value 所組成。

key 不能夠重複，否則會被後面的結果蓋過去。

可輸入 key 找尋您要找出來的值。

如果輸入的 key 不存在，那就會出現錯誤訊息。

key 只能是 字串 數字 tuple

```python
dict1 = {"a":100,"b":200,"c":300}
print(dict1)
print(dict1["b"])
print(dict1["a"])
print(dict1["x"])#KeyError 因為x的key不存在所以出錯
```

```python
dict1 = {"a":100,"b":200,"c":300}
for k in dict1:
    print(k,dict1[k])
print()    
for v in dict1.values():
    print(v)
```

## 新增與修改
+ 新增一筆資料
  + 請將 key 與 value 儲存至一個變數內。
  + 透過 dict.update(新增的資料) 這個方法的方式就可以新增。
+ 元素可否變更：
  + 以 = 指派方式指派給 key 就可以變更資料

```python
dict2 = {"ken":100,'vivin':95,"lucy":72}
dict2.update({"gigi":88,"vivin":98,"lucy":61})
print(dict2)#key不在更新 key在修改
```

```python
dict2 = {"ken":100,'vivin':95,"lucy":72}
print(dict2)
dict2["vivin"] = 77
print(dict2)
dict2["ben"] = 82
print(dict2)
```

```python
dict2 = {"ken":100,'vivin':95,"lucy":72}
print(dict2["vivin"])
k = "joy"
if k in dict2:
    print(dict2[k])
else:
    print(f"key:{k}不存在於dict2")
```

```python
dict2 = {"ken":100,'vivin':95,"lucy":72}
print(dict2.get("ken"))
print(dict2.get("joy",-1))#可以給予預設值 不會拋出例外
```

## del 

del dict[key]    刪除某一個 key 的資料

dict.clear( )     清除所有項目

del dict           刪除字典

```python
dict3 = {"ken":100,'vivin':95,"lucy":72}
print(dict3)
del dict3["vivin"]
print(dict3)
dict3.clear()
print(dict3)
```

```python
dict3 = {"a":100,"b":200,"c":300}
delKey = "y"
if delKey in dict3:
    del dict3[delKey]#如果能刪除 顯示已刪除 真的把數值移除
    print("已刪除")
else:
    print("無法刪除")
print(dict3)    
#如果不能刪除 顯示無法刪除
```

## 自訂函式的運用

#### 程式設計時儘量減少相同語法的重複出現，為什麼?

修改時可能忘記其他地方沒有改，導致程式執行出現錯誤結果。

減少重複語法對於之後的維修會比較方便。


#### 要如何減少重複的相同語法?

把重複的相同語法當作工具看待，需要的時候呼叫他。

這種工具於程式語法內叫做函數(function)。


```python
def func1():
    print("呼叫函式")
func1()
func1()  
```

```python
def func2(a,b):
    c = a + b
    print(c)
func2(2,5)
func2(6,9)  
```
