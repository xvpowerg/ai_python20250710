```python
import  matplotlib.pyplot as plt
import  numpy as np
city=['Delphi','Beijing','Washington','Tokyo','Moscow']
pos=np.arange(len(city))
gender=['Male','Female']
happing_index_male=[60,40,70,65,85]
happing_index_female=[30,60,70,55,75]
#bottom=happing_index_femalehapping_index_femaled
plt.bar(pos,happing_index_male,color="blue",bottom=happing_index_female)
plt.bar(pos,happing_index_female,color="pink")

plt.xticks(pos,city) 
plt.xlabel('capital')
plt.ylabel('index',rotation=0)
plt.title('barchart')
plt.legend(gender,loc=0)

plt.show()
```

```python
import  matplotlib.pyplot as plt
import  numpy as np
city=['Delphi','Beijing','Washington','Tokyo','Moscow']
pos=np.arange(len(city))
gender=['Male','Female']
happing_index_male=[60,40,70,65,85]
happing_index_female=[30,60,70,55,75]
bar_width = 0.35
plt.bar(pos,happing_index_male,bar_width,color="blue")
plt.bar(pos+bar_width,happing_index_female,bar_width,color="pink")

plt.xticks(pos,city) 
plt.xlabel('capital')
plt.ylabel('index',rotation=0)
plt.title('barchart')
plt.legend(gender,loc=1)

plt.show()
```

##  Scatter

```python
import matplotlib.pyplot as plt
import numpy as np
h=np.array([159,169,174,165,163,155,180,190])
w=np.array([55,45,70,60,65,45,72,50])
print(":",np.mean(h))
print(":",np.median(h))
print(":",np.mean(w))
print(":",np.median(w))

plt.scatter(h,w)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
h=np.array([159,169,174,165,163,155,180,190])
w=np.array([55,45,70,60,65,45,72,50])
print(h)
print(w)
plt.scatter(h,w)
plt.xlabel('h')
plt.ylabel('w')
for temp1 in zip(h,w):
  print(temp1)
  plt.text(temp1[0],temp1[1],temp1,fontsize=10)
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
values = [60, 80, 90, 55, 10, 30]
colors = ['b', 'g', 'r', 'c', 'm', 'y']
labels = ['US', 'UK', 'India', 'Germany', 'Australia', 'South Korea']
explode = (0.2, 0, 0, 0, 0, 0)
#counterclock
plt.pie(values, colors=colors, labels=values, explode=explode, counterclock=False, shadow=True)
plt.title('Population Density Index')
plt.legend(labels, loc=3)
plt.show( )
```
