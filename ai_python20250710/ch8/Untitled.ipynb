## 輸入四個月的支出金額後列出最多與最少的支出金額

```python
priceList = []
priceSum = 0
for i in range(1,5):
    p = int(input(f"請輸入{i}個月的支出金額"))
    priceSum += p
    priceList.append(p)
priceList.sort()
print("max:",priceList[-1])
print("min:",priceList[0])
print("Sum:",priceSum)
print(priceList)
```

## 匿名函數
不需要定義函數名稱，只需要用到運算式或表達分析語法。

Python使用lambda語法定義匿名函數。

匿名函數是一個表達式或計算式，並不是一個執行流程區塊。

匿名函數出現在一般函數不允許的地方，例如list裡面或著函數呼叫參數的位置。


```python
def f1(x,y,z):
    return x + y + z
print(f1(5,2,3))    
```

```python
print(f1,type(f1))
```

```python
myFuncx = f1#python內函式可被變數傳遞
myFuncx(5,6,7)
```

```python
def myMap(data):
    result = []
    for v in data:
        result.append(str(v))
    return result
values = [25,36,77]    

strValue = myMap(values)
msg = ",".join(strValue)
print(msg)
print(strValue)    
```

```python
def myMap(func, data):
    result = []
    for v in data:
        result.append(func(v))
    return result
def pow(n):    
    return n**2
values = [5,4,2,8]
powList = myMap(pow,values)
print(powList)
```

```python
myfunc2 = lambda n : n + 2
print(myfunc2,type(myfunc2))
```

```python
def myMap(func, data):
    result = []
    for v in data:
        result.append(func(v))
    return result
dataList = [5,2,3,7]    
ans =  myMap(myfunc2,dataList)    
print(ans)
ans2 = myMap(lambda x:x*3,dataList)
print(ans2)
```

```python
def f1(x):
    return x ** 2
def f2(x):
    return x ** 3
def f3(x):    
    return x ** 4
list1 = [f1,f2,f3]    
for f in list1:
    print(f(3))
```

```python
list2 = [lambda x:x ** 2,lambda x:x**3,lambda x:x**4]
for f in list2:
    print(f(3))
```

## Map

```python
myList = [1,2,3,'',4,5,'']
mapObj = map(lambda x : x if x != '' else 0,myList) 
print(list(mapObj))
```

```python
list4 = ["M","W","M","M","W"]
#M 轉換為0 W 轉換為 1
mapObj = map(lambda x : 0 if x == "M" else 1,list4)
print(list(mapObj))
```

```python
list4 = ["M","W","M","M","W"]
dict1 = {"M":0,"W":1}
mapObject = map(lambda x : dict1[x],list4)
print(list(mapObject))
```

## Filter

```python
a = list(range(1,21))
print(a)
newList = []
#產生一個新的List包含a內容的所有偶數
for v in a:
    if v % 2 == 0:
        newList.append(v)
print(newList)        
```

```python
a = list(range(1,21))
filterObj = filter(lambda x : x % 2 == 0,a)
print(list(filterObj))
```

```python
num = 10
num2 = 0
nums = [1,2,3,4]
print("Start")
try:
    print(num / num2)
except:
    print("錯誤發生了!")
print("End")
```

```python
num = 10
num2 = 1
nums = [1,2,3,4]
print("Start")

try:
    print(num / num2)
    print(nums[9])
except ZeroDivisionError:
    print("錯誤發生了~分母不可為零!")
except IndexError:
    print("錯誤發生了~Index不正確!")
print("End")
```

```python
num = 10
num2 = 1
nums = [1,2,3,4]
print("Start")

try:
    print(num / num2)
    print(nums[9])
except ZeroDivisionError:
    print("錯誤發生了~分母不可為零!")
except IndexError as ex:
    print("錯誤發生了~Index不正確!:",ex)
print("End")
```
