## 使用raise拋例外

```python
def testScore(score):
    if score <0 or score > 100:
        raise OverflowError("錯誤成績 必須0~100之間")
    else:
        print(score)
## 當    score < 0 或 score > 100  顯示錯誤成績 否則 顯示score的數值
try:
    testScore(-31)
    testScore(25)
    testScore(60)
except OverflowError as ex:
    print(ex)

```

## split 方法
+ split 方法能以比對規則針對字串進行切割後產生 list。
+ split 它的使用形式如下：
+ split(字串,最大分割次數)
   + 若不指定最大分割次數則全部分割

```python
import re
text = 'one, two...ten'
re1 = re.split("[,. ]+",text)
print(re1)
```
