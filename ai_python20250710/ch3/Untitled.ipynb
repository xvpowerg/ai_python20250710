## BMI分析作業


```python
h = input("請輸入身高單位公分")
w = input("請輸入身高單位公斤")
h = int(h)
w = int(w)
bmi = w / (h / 100) ** 2
print(bmi)
print(round(bmi,5))
print(f"bmi:{bmi:.2f}")
```

## FOR迴圈
### 知道數量時使用for loop
### python的for 有一種由頭到取數值的

想法


```python
a1 = range(10)
print(a1)
print(list(a1))
```

```python
a2 = range(2,8)
print(list(a2))
```

```python
a3 = range(2,18,3)
print(list(a3))
```

```python
a4 = range(5,-1,-1)
print(list(a4))
```

```python
for i in range(5):
    print(i,end = " ")
```

```python
for i in range(2,10):
    print(i,end=" ")
```

```python
for i in range(2,18,3):
    print(i,end= " ")
```

```python
for y in range(10,1,-2):
    print(y,end=" ")
```

```python
for i in range(1,6):
    print(i,end="")
    if i < 5:
        print(",",end="")        
```

## While

```python
a = 6
while a > 0:
    print(a)
    a -=  1
```

```python
n = "y"
while n == "y":
    a = int(input("請輸入第一個數字"))
    b = int(input("請輸入第一個數字"))
    ans = a * b
    print(a,"*",b,"=",ans)
    n = input("是否繼續(y/n)")    
```

```python
for i in range(1,11):
    print(i,end=" ")
    if i == 5:
        break#離開迴圈
```

```python
for i in range(1,6):
    print("start:",i)
    if i == 2 or i == 4:
        continue
    print("end:",i)
    print("===================")
```

```python
#1 2 3 4 6 8 9 10
for i in range(1,11):
    if i == 5 or i == 7:
        continue
    print(i,end=" ")
```

```python
print("請輸入大於80的整數")
while True:
    a = int(input("請輸入數字:"))
    if a > 80:
        print("a=",a)
        break
```
