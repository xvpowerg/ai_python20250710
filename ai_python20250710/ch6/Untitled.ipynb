```python
txt = input("")
strList = []
strList.append(txt)
msg = txt
inputMsg = f"-{txt[-1]}-"
count = 0
while count < 5:
    tmpTxt = input(inputMsg)
    if tmpTxt[0] == txt[-1]:
        txt = tmpTxt
        strList.append(tmpTxt)
        msg = "-".join(strList)
        inputMsg = f"-{txt[-1]}-"
    else:
        inputMsg = f"-{txt[-1]}-"
        count += 1
    print(":",msg)    
```
