# List VS Tuple
+ tuple 與 list 的差異
  + 括號不同：
    + tuple 是用小括號將元素包起來。
    + list 是用中括號將元素包起來。
+ 元素可否變更：
   + tuple 決定了元素的內容就不可以變更。
   + list 可以隨時修改

```python
list1 = ["A","B"]
tuple1 = ("X","Y")
print(list1)
print(tuple1)
print(list1[1])
print(tuple1[0])
print("====================")
list1[1] = "G"
print(list1[1])
tuple1[0] = ""
```

```python
list1 = ["A","B","C","D","E"]
print(list1[1:3])
print(list1[-2:])
print(list1[::-1])
print(list1[-3::-1])
```

```python
myList = ["A"]
print(type(myList))
myTuple = ("A",)
print(type(myTuple))
```

```python
list1 = [1,2,3,4,5]
list2 = [21,22,23,24,25]
print(list1)
print(len(list1))
list1.append(100)
list1.append(105)
list1.append(106)
print(list1)
print(len(list1))
list1.append(list2)
print(list1)
print(len(list1))
```

```python
list1 = [1,2,3,4,5]
list2 = [21,22,23,24,25]
list1 += list2
print(list1)
```

```python
list1 = [1,2,3,4,5]
list2 = [21,22,23,24,25]
list1.extend(list2)
print(list1)
```

## list 項目插入新資料
+ list 可使用 insert 方式插入新的資料：
   + list.insert(索引值，插入的資料)。
   + 若索引值超出範圍則會改為 append 方式附加。

```python
list1 = ["a","b","c","d","e"]
print(list1)
list1.insert(1,"X")
print(list1)
list1.insert(-2,"Y")
print(list1)
list1.insert(99,"Z")
print(list1)
```

```python
# Set
myList = ["a","b","a","a","c","b","c"]
print(myList)
mySet = set(myList)
print(mySet)#過濾重複
```

```python
list1 = ["a","b","a","c","b"]
filterList = []
for v in list1:
    if v not in filterList:
        filterList.append(v)
print(filterList)
```

## 移除 
+ list 可透過以下三種方式移除項目：
  + list.remove( )
  + list.pop( )
  + del list[ ]

```python
list1 = ["1","x","2","x","3"]#左往右找到的先移除
list1.remove("x")
print(list1)
list1.remove("x")
print(list1)
if "x" in list1:
    list1.remove("x")
```

```python
list1 = ["1","x","2","x","3"]
#用迴圈 全刪!
key = "x"
while  True:
    if key not in list1:
        break
    list1.remove(key)     
print(list1)
```

```python
list1 = ["1","x","2","x","3"]
key = "x"
while key in list1:
    list1.remove(key)
print(list1)    
```

### pop

如果 pop( ) 內沒有參數，則移除最後一筆資料。

pop() 內若有參數，則移除指定位置的資料。

```python
list2 = []
list2.append("Ken")
list2.append("Vivin")
list2.append("Lindy")
list2.append("Lucy")
print(list2)
v1 = list2.pop()
print(v1)
print(list2)
v1 = list2.pop(0)
print(v1)
print(list2)
```

```python
list2 = []
list2.append("Ken")
list2.append("Vivin")
list2.append("Lindy")
list2.append("Lucy")

while len(list2):
    print(list2.pop())
    
```

### Del
+ list 可透過 del 方式移除指定位置資料。
+ del list[ ] 必須指定範圍：
  + del list[n] 代表移除第 n 位索引值資料。
  + del list[:m] 代表移除第 m 位索引值之前 ( 不包含 m ) 資料。
  + del list[n:m] 代表移除第 n 到 m 位索引值之前 ( 不包含 m ) 資料。

```python
list2 = ["A","B","C","D","E"]
print(list2)
del list2[1]
print(list2)
del list2[1:3]
print(list2)
```
