## BMI分析作業


```python
h = input("請輸入身高")
w = input("請輸入體重")
h = int(h)
w = int(w)
bmi = w / (h/100) ** 2
print(bmi)
if  0 < bmi <= 18.5:
    print("過輕")
elif 18.5 < bmi <= 25:
    print("正常")
elif 25 < bmi <= 30:
    print("過重")
elif bmi > 30:
    print("胖胖")
else:
    print("error")
    
```

## find

字串可以搜尋特定內容的位置。

搜尋的字串位置編號由 0 開始。

字串.find(搜尋字串,起始位置,結束位置)

搜尋時不包含結束位置。

若搜尋不到則回傳 -1。

若沒有起始位置，也沒有結束位置，預設由 0 開始。

若沒有結束位置，預設到最後一個字。


```python
str1 = "This is Python.That is Java,This is SQLite"
print(str1)
search1 = "Python"
print(str1.find(search1))
search1 = "not"
print(str1.find(search1))#找不到一定是-1
search1 = "is"
print(str1.find(search1))
search1 = "is"
print(str1.find(search1,4))
```

```python
str1 = "This is Python.That is Java,This is SQLite"
search1 = "is"
i = 0
i = str1.find(search1,i)
print(i)
i = str1.find(search1,i+1)
print(i)
i = str1.find(search1,i+1)
print(i)
```

```python
str1 = "This is Python.That is Java,This is SQLite"
search1 = "is"
index = 0 
while True:
    index = str1.find(search1,index)
    if index == -1:
        break
    print("index:",index)    
    index += 1    
```

## join

join( ) : 於字串或著字元之間加入指定的文字

指定文字.join(要插入的字串)

```python
str1 = "ABCDE"
print("len:",len(str1))
newStr = ""
for i in range(len(str1)):
    #print(i,str1[i])
    newStr += str1[i] +"-"
print(newStr[:-1])    
```

```python
newStr = "-".join("ABCDE")
print(newStr)
```
